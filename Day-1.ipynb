# Loops - For

```python
lst = [1,2,3,4,5,6,7,8]
for element in lst:
    print(element)
```

```python
for indx,value in enumerate(lst):
    print(f'{indx}    -    {value}')
```

```python
def stats(nums,label="data"):
    return {
        "label":label,
        "avg":sum(nums)/len(nums),
        "min":min(nums),
        "max":max(nums)
    }
```

```python
result=stats([15,51,16,61],label='scores')
```
